```python
import os
from PIL import Image
from tqdm import tqdm_notebook as tqdm
import piexif
from random import shuffle
import shutil
```

```python
def filter_bad_images(directory):
    for (dirpath, dirnames, filenames) in os.walk(directory):
        #print(dirpath, dirnames)
        for fname in tqdm(filenames):
            full_path = os.path.join(dirpath, fname)
            try:
                piexif.remove(full_path)
            except Exception as err:
                print(f"bad exif data, deleting {full_path}")
                os.remove(full_path)
                continue
            if fname[0] == '.':
                os.remove(full_path)
                print(f"file deleted because of wierd name {full_path}")
            try:
                Image.open(full_path)
            except Exception as err:
                print(f"deleting file coz i can't open it {full_path}")
                os.remove(full_path)
    
```

```python
filter_bad_images('/home/ec2-user/data/train/')
```

```python
def create_validation(src_dir, dest_dir, num=4000):
    all_files = []
    for (dirpath, dirnames, filenames) in os.walk(src_dir):
        for fname in filenames:
            full_path = os.path.join(dirpath, fname)
            all_files.append((full_path, fname))
    shuffle(all_files)
    for i in range(num):
        path, fname = all_files[i]
        print(f"moving: {path}")
        shutil.move(path, os.path.join(dest_dir, fname))
    
```

```python
create_validation('/home/ec2-user/data/train/imagenet', '/home/ec2-user/data/test/imagenet')
```
